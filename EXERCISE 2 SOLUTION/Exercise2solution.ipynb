```python
sql_commands = """

USE stage;

CREATE TABLE IF NOT EXISTS reviews (
    id VARCHAR(10) PRIMARY KEY,
    Title VARCHAR(600),
    User_id VARCHAR(25),
    profileName VARCHAR(255),
    helpful_votes INT,
    total_votes INT,
    review_score INT CHECK (review_score >= 0 AND review_score <= 5),
    review_time INT,
    review_summary TEXT,
    review_text TEXT
);

CREATE TABLE IF NOT EXISTS book_details (
    Title VARCHAR(700) PRIMARY KEY,
    Description TEXT,
    image VARCHAR(255),
    previewLink VARCHAR(1012),
    publisher VARCHAR(255),  -- Publisher included directly
    publishedDate DATE,
    infoLink VARCHAR(512),
    ratingsCount INT
);

CREATE TABLE IF NOT EXISTS authors (
    author_id INT AUTO_INCREMENT PRIMARY KEY,
    author_name VARCHAR(255) NOT NULL
);

CREATE TABLE IF NOT EXISTS categories (
    category_id INT AUTO_INCREMENT PRIMARY KEY,
    category_name VARCHAR(255) NOT NULL
);

CREATE TABLE IF NOT EXISTS book_authors (
    Title VARCHAR(700),
    author_id INT,
    PRIMARY KEY (Title, author_id),
    FOREIGN KEY (Title) REFERENCES book_details(Title),
    FOREIGN KEY (author_id) REFERENCES authors(author_id)
);

CREATE TABLE IF NOT EXISTS book_categories (
    Title VARCHAR(700),
    category_id INT,
    PRIMARY KEY (Title, category_id),
    FOREIGN KEY (Title) REFERENCES book_details(Title),
    FOREIGN KEY (category_id) REFERENCES categories(category_id)
);

CREATE TABLE IF NOT EXISTS reviews_log (
    log_id INT AUTO_INCREMENT PRIMARY KEY,
    action VARCHAR(10),
    review_id VARCHAR(10),
    Title VARCHAR(600),
    review_time INT
);

-- Use a delimiter for the trigger definitions
DELIMITER //

-- Create an after insert trigger on reviews
CREATE TRIGGER after_insert_review
AFTER INSERT ON reviews
FOR EACH ROW
BEGIN
    INSERT INTO reviews_log (action, review_id, Title, review_time)
    VALUES ('INSERT', NEW.id, NEW.Title, NEW.review_time);
END //

-- Create an after update trigger on reviews
CREATE TRIGGER after_update_review
AFTER UPDATE ON reviews
FOR EACH ROW
BEGIN
    INSERT INTO reviews_log (action, review_id, Title, review_time)
    VALUES ('UPDATE', NEW.id, NEW.Title, NEW.review_time);
END //

-- Create an after delete trigger on reviews
CREATE TRIGGER after_delete_review
AFTER DELETE ON reviews
FOR EACH ROW
BEGIN
    INSERT INTO reviews_log (action, review_id, Title, review_time)
    VALUES ('DELETE', OLD.id, OLD.Title, OLD.review_time);
END //

DELIMITER ;  -- Reset the delimiter
"""

# Write the SQL commands to a file
with open('create_tables_stage.sql', 'w') as f:
    f.write(sql_commands)
```

```python
stage_connection = mysql.connector.connect(
    host='localhost',
    user='admin',
    password='admin_password',
    database='stage'
)
```

```python
stage_cursor = stage_connection.cursor()
```

```python
procedure = """

CREATE PROCEDURE transfer_reviews_to_stage()
BEGIN
    DECLARE done INT DEFAULT FALSE;
    DECLARE batch_size INT DEFAULT 100;
    DECLARE offset INT DEFAULT 0;

    -- Loop to process batches
    WHILE NOT done DO
        -- Insert records in batches
        INSERT INTO stage.reviews (
            id,
            Title,
            User_id,
            profileName,
            helpful_votes,
            total_votes,
            review_score,
            review_time,
            review_summary,
            review_text
        )
        SELECT
            r.id,
            r.Title,
            r.User_id,
            r.profileName,
            CASE
                WHEN r.review_helpfulness = '' THEN 0
                WHEN LOCATE('/', r.review_helpfulness) > 0 THEN
                    CAST(SUBSTRING_INDEX(r.review_helpfulness, '/', 1) AS UNSIGNED)
                ELSE 0
            END AS helpful_votes,
            CASE
                WHEN r.review_helpfulness = '' THEN 0
                WHEN LOCATE('/', r.review_helpfulness) > 0 THEN
                    CAST(SUBSTRING_INDEX(r.review_helpfulness, '/', -1) AS UNSIGNED)
                ELSE 0
            END AS total_votes,
            r.review_score,
            r.review_time,
            r.review_summary,
            r.review_text
        FROM raw.reviews r
        WHERE r.id IS NOT NULL
        LIMIT batch_size OFFSET offset
        LOCK IN SHARE MODE;  -- Added locking mode for concurrent reads

        -- Check if any rows were inserted
        IF ROW_COUNT() < batch_size THEN
            SET done = TRUE;  -- Exit the loop if fewer rows are inserted (end of data)
        END IF;

        -- Increment the offset for the next batch
        SET offset = offset + batch_size;

    END WHILE;
END;
"""

# Execute the procedure creation in the stage database
stage_cursor.execute(procedure)
stage_connection.commit()

# Call the stored procedure to transfer the data
try:
    stage_cursor.execute("CALL transfer_reviews_to_stage();")
    stage_connection.commit()
except Exception as e:
    print(f"An error occurred: {e}")
    stage_connection.rollback()  # Rollback in case of any error
```

```python
procedure = """
CREATE PROCEDURE transfer_book_details_to_stage()
BEGIN
    DECLARE done INT DEFAULT 0;
    DECLARE batch_size INT DEFAULT 100;
    DECLARE offset INT DEFAULT 0;
    DECLARE row_count INT DEFAULT 0;

    WHILE done = 0 DO
        -- Insert in batches
        INSERT INTO stage.book_details (
            Title,
            Description,
            image,
            previewLink,
            publisher,
            publishedDate,
            infoLink,
            ratingsCount
        )
        SELECT
            b.Title,
            b.Description,
            b.image,
            b.previewLink,
            b.publisher,
            CASE
                -- Convert YYYY to YYYY-01-01
                WHEN LENGTH(TRIM(b.publishedDate)) = 4 AND TRIM(b.publishedDate) REGEXP '^[0-9]{4}$' THEN
                    STR_TO_DATE(CONCAT(TRIM(b.publishedDate), '-01-01'), '%Y-%m-%d')

                -- Convert YYYY* to YYYY-01-01 (removing the asterisk)
                WHEN LENGTH(TRIM(b.publishedDate)) = 5 AND TRIM(b.publishedDate) LIKE '____*' THEN
                    STR_TO_DATE(CONCAT(SUBSTRING(TRIM(b.publishedDate), 1, 4), '-01-01'), '%Y-%m-%d')

                -- Convert YYYY-MM-DD format
                WHEN LENGTH(TRIM(b.publishedDate)) = 10 AND TRIM(b.publishedDate) REGEXP '^[0-9]{4}-[0-9]{2}-[0-9]{2}$' THEN
                    STR_TO_DATE(TRIM(b.publishedDate), '%Y-%m-%d')

                -- Handle ISO 8601 formats
                WHEN TRIM(b.publishedDate) LIKE '%T%' THEN
                    STR_TO_DATE(SUBSTRING(TRIM(b.publishedDate), 1, 10), '%Y-%m-%d')

                -- Return NULL for any other formats
                ELSE NULL
            END AS publishedDate,
            b.infoLink,
            b.ratingsCount
        FROM raw.book_details b
        WHERE
            (CASE
                -- Ensure only valid conversions are processed
                WHEN LENGTH(TRIM(b.publishedDate)) = 4 AND TRIM(b.publishedDate) REGEXP '^[0-9]{4}$' THEN
                    TRUE
                WHEN LENGTH(TRIM(b.publishedDate)) = 5 AND TRIM(b.publishedDate) LIKE '____*' THEN
                    TRUE
                WHEN LENGTH(TRIM(b.publishedDate)) = 10 AND TRIM(b.publishedDate) REGEXP '^[0-9]{4}-[0-9]{2}-[0-9]{2}$' THEN
                    TRUE
                WHEN TRIM(b.publishedDate) LIKE '%T%' THEN
                    TRUE
                ELSE FALSE
            END)
        LIMIT batch_size OFFSET offset;

        -- Update the offset for the next batch
        SET offset = offset + batch_size;

        -- Check how many rows were inserted
        SET row_count = ROW_COUNT();
        IF row_count = 0 THEN
            SET done = 1;
        END IF;
    END WHILE;
END;


"""

# Execute the procedure creation in the stage database
stage_cursor.execute(procedure)
stage_connection.commit()

# Call the stored procedure to transfer the data
try:
    stage_cursor.execute("CALL transfer_book_details_to_stage();")
    stage_connection.commit()
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
procedure = """
CREATE PROCEDURE transfer_authors_to_stage()
BEGIN
    DECLARE done INT DEFAULT 0;
    DECLARE author_name VARCHAR(255);
    DECLARE book_title VARCHAR(700);
    DECLARE book_id INT;

    DECLARE cur CURSOR FOR
        SELECT Title FROM raw.book_details WHERE authors IS NOT NULL;

    DECLARE CONTINUE HANDLER FOR NOT FOUND SET done = 1;

    OPEN cur;

    read_loop: LOOP
        FETCH cur INTO book_title;

        IF done THEN
            LEAVE read_loop;
        END IF;

        SELECT book_id INTO book_id
        FROM stage.book_details
        WHERE Title LIKE CONCAT('%', book_title, '%') LIMIT 1;

        IF book_id IS NULL THEN
            LEAVE read_loop;
        END IF;

        SELECT authors INTO @authors_list FROM raw.book_details WHERE Title = book_title;

        IF @authors_list IS NULL OR @authors_list = '' THEN
            LEAVE read_loop;
        END IF;

        WHILE LOCATE(',', @authors_list) > 0 DO
            SET author_name = TRIM(SUBSTRING_INDEX(@authors_list, ',', 1));
            SET @authors_list = SUBSTRING(@authors_list FROM LOCATE(',', @authors_list) + 1);

            INSERT INTO stage.authors (author_name)
            VALUES (author_name)
            ON DUPLICATE KEY UPDATE author_id = LAST_INSERT_ID(author_id);

            INSERT INTO stage.book_authors (book_id, author_id)
            VALUES (book_id, LAST_INSERT_ID());
        END WHILE;

        SET author_name = TRIM(@authors_list);
        IF author_name IS NOT NULL AND author_name != '' THEN
            INSERT INTO stage.authors (author_name)
            VALUES (author_name)
            ON DUPLICATE KEY UPDATE author_id = LAST_INSERT_ID(author_id);

            INSERT INTO stage.book_authors (book_id, author_id)
            VALUES (book_id, LAST_INSERT_ID());
        END IF;

    END LOOP;

    CLOSE cur;
END;


"""

# Execute the procedure creation in the stage database
stage_cursor.execute(procedure)
stage_connection.commit()

# Call the stored procedure to transfer the data
try:
    stage_cursor.execute("CALL transfer_authors_to_stage();")
    stage_connection.commit()
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
procedure = """
CREATE PROCEDURE transfer_categories_to_stage()
BEGIN
    DECLARE done INT DEFAULT 0;
    DECLARE category_name VARCHAR(255);
    DECLARE book_title VARCHAR(700);
    DECLARE categories_list VARCHAR(1000);
    DECLARE book_id INT;

    -- Cursor for iterating over book titles and their categories
    DECLARE cur CURSOR FOR
        SELECT Title, categories FROM raw.book_details WHERE categories IS NOT NULL;

    -- Handler for the end of the cursor
    DECLARE CONTINUE HANDLER FOR NOT FOUND SET done = 1;

    OPEN cur;

    read_loop: LOOP
        FETCH cur INTO book_title, categories_list;

        -- Exit the loop if there are no more rows to fetch
        IF done THEN
            LEAVE read_loop;
        END IF;

        -- Trim the book title for better matching
        SET book_title = TRIM(book_title);

        -- Fetch the book_id based on the book title (with LIKE for potential matching issues)
        SELECT book_id INTO book_id FROM stage.book_details WHERE Title LIKE CONCAT('%', book_title, '%') LIMIT 1;

        -- Proceed only if a valid book_id was retrieved
        IF book_id IS NOT NULL THEN
            -- Process each category in the list
            WHILE LENGTH(categories_list) > 0 DO
                -- Get the first category
                SET category_name = TRIM(SUBSTRING_INDEX(categories_list, ',', 1));

                -- Remove the first category from categories_list
                SET categories_list = TRIM(SUBSTRING(categories_list FROM LOCATE(',', categories_list) + 1));

                -- Check if the category_name is not empty
                IF category_name != '' THEN
                    -- Insert the category if it doesn't exist, otherwise get the existing category_id
                    INSERT INTO stage.categories (category_name)
                    VALUES (category_name)
                    ON DUPLICATE KEY UPDATE category_id = LAST_INSERT_ID(category_id);

                    -- Link the category to the book in the junction table using the book_id
                    INSERT INTO stage.book_categories (book_id, category_id)
                    VALUES (book_id, LAST_INSERT_ID());
                END IF;
            END WHILE;
        END IF;
    END LOOP;

    CLOSE cur;
END;

"""

# Execute the procedure creation in the stage database
stage_cursor.execute(procedure)
stage_connection.commit()

# Call the stored procedure to transfer the data
try:
    stage_cursor.execute("CALL transfer_categories_to_stage();")
    stage_connection.commit()
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
sql_commands = """
USE hist;

CREATE TABLE IF NOT EXISTS hist.authors (
    author_id INT PRIMARY KEY,
    author_name VARCHAR(255) NOT NULL,
    start_date DATE,
    end_date DATE,
    current_flag BOOLEAN
);

-- Dimensional Table for Books
CREATE TABLE IF NOT EXISTS hist.books (
    book_id INT AUTO_INCREMENT PRIMARY KEY,
    Title VARCHAR(700),
    Description TEXT,
    image VARCHAR(255),
    previewLink VARCHAR(1012),
    publisher VARCHAR(255),
    publishedDate DATE,
    infoLink VARCHAR(512),
    ratingsCount INT,
    start_date DATE,
    end_date DATE,
    current_flag BOOLEAN
);

-- Dimensional Table for Categories
CREATE TABLE IF NOT EXISTS hist.categories (
    category_id INT PRIMARY KEY,
    category_name VARCHAR(255) NOT NULL,
    start_date DATE,
    end_date DATE,
    current_flag BOOLEAN
);

-- Junction Table for Book-Authors Relationship
CREATE TABLE IF NOT EXISTS hist.book_authors (
    book_id INT,
    author_id INT,
    PRIMARY KEY (book_id, author_id),
    FOREIGN KEY (book_id) REFERENCES hist.books(book_id),
    FOREIGN KEY (author_id) REFERENCES hist.authors(author_id)
);

-- Junction Table for Book-Categories Relationship
CREATE TABLE IF NOT EXISTS hist.book_categories (
    book_id INT,
    category_id INT,
    PRIMARY KEY (book_id, category_id),
    FOREIGN KEY (book_id) REFERENCES hist.books(book_id),
    FOREIGN KEY (category_id) REFERENCES hist.categories(category_id)
);

-- Fact Table for Reviews
CREATE TABLE IF NOT EXISTS hist.reviews (
    review_id INT AUTO_INCREMENT PRIMARY KEY,
    book_id INT,
    user_id VARCHAR(25),
    profile_name VARCHAR(255),
    helpful_votes INT,
    total_votes INT,
    review_score INT CHECK (review_score >= 0 AND review_score <= 5),
    review_time DATE,
    review_summary TEXT,
    review_text TEXT,
    start_date DATE,
    end_date DATE,
    current_flag BOOLEAN,
    FOREIGN KEY (book_id) REFERENCES hist.books(book_id)
);
"""

# Write the SQL commands to a file
with open('create_tables_hist.sql', 'w') as f:
    f.write(sql_commands)
```

```python
# Connect to the MySQL database
db_connection_hist = mysql.connector.connect(
    host='localhost',
    user='admin',
    password='admin_password',
    database='hist'
)

db_cursor_hist = db_connection_hist.cursor()

# SQL statements for creating the stored procedures
create_procedure_insert_new_review = """

CREATE PROCEDURE insert_new_review(
    IN p_book_id INT,
    IN p_user_id VARCHAR(25),
    IN p_profile_name VARCHAR(255),
    IN p_helpful_votes INT,
    IN p_total_votes INT,
    IN p_review_score INT,
    IN p_review_time DATE,
    IN p_review_summary TEXT,
    IN p_review_text TEXT
)
BEGIN
    -- Enforce the check constraint within the procedure
    IF p_review_score < 0 OR p_review_score > 5 THEN
        SIGNAL SQLSTATE '45000' SET MESSAGE_TEXT = 'Review score must be between 0 and 5.';
    ELSE
        INSERT INTO hist.reviews (
            book_id, user_id, profile_name, helpful_votes,
            total_votes, review_score, review_time,
            review_summary, review_text, start_date, current_flag
        ) VALUES (
            p_book_id, p_user_id, p_profile_name, p_helpful_votes,
            p_total_votes, p_review_score, p_review_time,
            p_review_summary, p_review_text, NOW(), TRUE
        );
    END IF;
END;

"""

create_procedure_update_book_average_rating = """

CREATE PROCEDURE update_book_average_rating(IN p_book_id INT)
BEGIN
    DECLARE avg_rating DECIMAL(3, 2);

    SELECT AVG(review_score) INTO avg_rating
    FROM hist.reviews
    WHERE book_id = p_book_id;

    UPDATE hist.books
    SET ratingsCount = (SELECT COUNT(*) FROM hist.reviews WHERE book_id = p_book_id),
        avg_rating = avg_rating -- Assuming there's an `avg_rating` column in `hist.books`
    WHERE book_id = p_book_id;
END;
"""

create_procedure_delete_old_reviews = """

CREATE PROCEDURE delete_old_reviews(IN p_days INT)
BEGIN
    DELETE FROM hist.reviews
    WHERE review_time < NOW() - INTERVAL p_days DAY;
END;
"""

create_procedure_update_review_score = """
CREATE PROCEDURE UpdateReviewScore(
    IN p_review_id INT,  -- Changed the parameter name to avoid confusion
    IN p_new_score DECIMAL(3, 2)
)
BEGIN
    -- Update the review score
    UPDATE fact_reviews
    SET review_score = p_new_score  -- Use the new score parameter
    WHERE review_id = p_review_id;   -- Use the review ID parameter

    -- Log the update in the reviews_log table
    INSERT INTO reviews_log (action, review_id, book_id, review_time, log_timestamp)
    SELECT 'UPDATE_SCORE', p_review_id, book_id, review_time, NOW()
    FROM fact_reviews
    WHERE review_id = p_review_id;   -- Use the review ID parameter
END;
"""


# Execute the stored procedure creation commands
try:
    db_cursor_hist.execute(create_procedure_insert_new_review)
    print("Stored Procedure 'insert_new_review' created successfully.")

    db_cursor_hist.execute(create_procedure_update_book_average_rating)
    print("Stored Procedure 'update_book_average_rating' created successfully.")

    db_cursor_hist.execute(create_procedure_delete_old_reviews)
    print("Stored Procedure 'delete_old_reviews' created successfully.")

    db_cursor_hist.execute(create_procedure_update_review_score)
    print("Stored Procedure 'update_review_score' created successfully.")

except mysql.connector.Error as err:
    print(f"Error: {err}")

db_cursor_hist.close()
db_connection_hist.close()
```
